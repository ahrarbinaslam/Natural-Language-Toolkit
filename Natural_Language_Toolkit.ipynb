# Removing Stopwords from Textual Data

```python
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
stop=stopwords.words('english')
print(stop)

text_stop="i am going to karachi."
from nltk.tokenize import word_tokenize
tokenized=word_tokenize(text_stop)
print(tokenized)

tokenize_words_without_stop_words=[]
for word in tokenized:
  if word not in stop:
    tokenize_words_without_stop_words.append(word)
print(tokenize_words_without_stop_words)

print(set(tokenized)-set(tokenize_words_without_stop_words))
```

```python
text_stop="This is my school, it is one of the best schools, you should've gone here, so we could have played in the garden"
from nltk.corpus import stopwords
stop_words=stopwords.words('english')
print("The stop words in the library are ",stop_words)

from nltk.tokenize import word_tokenize
tokenize_words=word_tokenize(text_stop)
print("Tokenized words", tokenize_words)

tokenize_words_without_stop_words=[]
for word in tokenize_words:
  if word not in stop_words:
    tokenize_words_without_stop_words.append(word)
print(tokenize_words_without_stop_words)

print("The words that have been removed are \n")
print(set(tokenize_words) - set(tokenize_words_without_stop_words))
```

# VADER Sentiment Analysis

```python
import nltk
nltk.download('vader_lexicon') #Lexicon is a dictionary of words related to any topic.
```

```python
sentence="don't can't"
from nltk.sentiment import SentimentIntensityAnalyzer
sia=SentimentIntensityAnalyzer()
sentiment_scores=sia.polarity_scores(sentence)
print(sentiment_scores)

#Roberta model
#Huggingface
```

```python
if sentiment_scores['compound']>=0.05:
  print("Positive Sentiment")
elif sentiment_scores['compound']<=-0.05:
  print("Negative Sentiment")
else:
  print("Neutral Sentiment")
```

# Generating a Basic Word Cloud

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

text = "Cristiano Ronaldo is a football legend that has managed to conquer the world of soccer and fashion as well. He was born in Portugal. He grew up in his hometown called Madeira and when he was twelve he moved to Lisbon. He always had a passion over football since his childhood. Ronaldo started his football career as a youth player for Andorinha, he played for two years there then later he went to C.D. Nacional. In (1997) he moved to Sporting CP Lisbon football Club. Later in 2003 he signed for United. Ronaldo improved a lot in there. He helped the club to wins a lot of trophies which made his named the European and worlds best player in 2008.ronaldo became the most expensive footballer when he made a transfer to Real Madrid in 2009. Until today he still plays for Read Madrid.he has won he world best player awards thice in his career so far."
wordcloud1 = WordCloud(background_color="white", height=800,width=1000, mode="RGBA").generate(text)

# Plot the word cloud using imshow, image show function
plt.figure(figsize=(10,5))
plt.imshow(wordcloud1, interpolation='bilinear') #By setting interpolation='bilinear', the word cloud image will be displayed with a more visually pleasing appearance.
plt.axis("off")
plt.savefig("abc.png", dpi=400)
plt.show()

#colormap=viridis
```

# Reading a File from Excel and Computing Word Clouds

```python
import pandas as pd
data= pd.read_excel('Book1.xlsx')
text_data=data['Text'].astype(str).str.cat(sep='')
```

```python
from wordcloud import WordCloud
from matplotlib import pyplot as plt
wordcloud1=WordCloud(background_color="white", width=1500, height=1000, mode="RGBA", colormap="viridis", include_numbers=True).generate(text_data)
plt.figure(figsize=(10,5))
plt.imshow(wordcloud1, interpolation='bilinear')
plt.axis('off')
plt.savefig("abc.jpg", dpi=400)
plt.show()


#Some of the colormap options are cool, hot, rainbow, jet, Blues, Greens, Reds, Oranges, Purples
```
